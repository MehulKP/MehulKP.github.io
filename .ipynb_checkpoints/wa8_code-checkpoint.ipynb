# Weekly Assignment #8: DPD Data

```python
import pandas as pd
import altair as alt
```

## Reading in the Denver policing data for November

```python
all_data_df = pd.read_csv("Denver policing data November 2020 .csv")
all_data_df
```

## Filters data by date

```python
precinct_gb = all_data_df.groupby(pd.Grouper(key='date'))
```

## Aggregates the data by the total number of arrests made each day of November

```python
agg_d = {'arrest_made':['sum']}
daily_arrests_df = precinct_gb.agg(agg_d)
daily_arrests_df = daily_arrests_df.sort_values(by='date')
daily_arrests_df
```

#### Data Manipulation [for formatting purposes]

```python
daily_arrests_df = daily_arrests_df.reset_index(level=[0])
dates_df = daily_arrests_df['date'].to_frame()
merged_daily_arrests_df = dates_df.join(daily_arrests_df['arrest_made']).rename(columns={"sum": "# of Arrests", "date": "Date"})
merged_daily_arrests_df
```

## Data Visualization

```python
alt.Chart(merged_daily_arrests_df).mark_bar().encode(y='# of Arrests', x='Date').properties(title='Denver Arrest Frequency in November of 2012')
```
